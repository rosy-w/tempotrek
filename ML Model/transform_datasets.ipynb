```python
csv_files=[]
csv_paths=[]

count=0
for file in os.listdir():
  path= os.listdir(file)
  folder_path= os.path.abspath(file)
  for name in path:
    if fnmatch.fnmatch(name,'*.csv'):
      full_path = folder_path+'/'+name 
      csv_files.append(name)
      csv_paths.append(full_path)
      count+=1

print(f"There are {count} csv files")
print(f"The list has {len(csv_files)} items")
```

#Accelerometer and compass CSV data

I had to create 2 arrays: the paths and filenames , to enable the creation of a csv file in a new directory.

```python
accel_files=[]
cmpss_files=[]

accel_paths=[]
cmpss_paths=[]

for file in csv_files:
    if fnmatch.fnmatch(file, '*accel*'):
      accel_files.append(file)
    elif fnmatch.fnmatch(file, '*cmpss*'):
      cmpss_files.append(file)

for file in csv_paths:
    if fnmatch.fnmatch(file, '*accel*'):
      accel_paths.append(file)
    elif fnmatch.fnmatch(file, '*cmpss*'):
      cmpss_paths.append(file) 
           
print(f'Total no. of Accel files : {len(accel_files)}')
print(f'Total no. of Compass files : {len(cmpss_files)}')

print(f'Accel files : {accel_files}')
print(f'Compass files : {cmpss_files}')

print(f'Accel paths : {accel_paths}')
print(f'Compass paths : {cmpss_paths}')
```

####Add 'Accel_' prefix to all columns



```python
def checkFilepath(path):
  import sys
  fail="File not found"

  try:
    df = pd.read_csv(path)
    return path
  except FileNotFoundError as e:
    return fail
```

```python
count=0
j=0

for fname in accel_files:
  count+=1

for fname in accel_paths: 
  if checkFilepath(fname) != "File not found":
    df = pd.read_csv(fname)
    df2 = pd.DataFrame()
    df2= df
    for col in df.columns: 
      if fnmatch.fnmatch(col,'*Accel*'):
        pass
      else:
        if fnmatch.fnmatch(col,'*time*'):
          colname = 'time'
          df2.rename(columns = {col:colname}, inplace = True)
        else:
          colname = 'Accel_'+ col
          df2.rename(columns = {col:colname}, inplace = True)
    
    # file name with extension
    file_name = os.path.basename(fname)

    # file name without extension
    newname=(os.path.splitext(file_name)[0])

    #Convert the dataframe to csv 
    df2.to_csv(tranformed_accel_dir+"/"+newname+".csv")
    j+=1

    print(f"{j} files transformed out of {count}")
```

The same process for compass csv files

```python
#df3 and df4
count=0
j=0

for fname in cmpss_files:
  count+=1

for fname in cmpss_paths: 
  if checkFilepath(fname) != "File not found":
    df3 = pd.read_csv(fname)
    df4 = pd.DataFrame()
    df4= df3
    for col in df3.columns: 
      if fnmatch.fnmatch(col,'*Cmpss*'):
        pass
      else:
        if fnmatch.fnmatch(col,'*time*'):
          colname = 'time'
          df4.rename(columns = {col:colname}, inplace = True)
        else:
          colname = 'Cmpss_'+ col
          df4.rename(columns = {col:colname}, inplace = True)
    
    # file name with extension
    file_name = os.path.basename(fname)

    # file name without extension
    newname=(os.path.splitext(file_name)[0])

    #Convert the dataframe to csv 
    df4.to_csv(tranformed_cmpss_dir+"/"+newname+".csv")
    j+=1

    print(f"{j} files transformed out of {count}")
```
